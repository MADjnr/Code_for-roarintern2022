Rose Curves

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 4.2 #Change the values of k to see how the number of the pedals vary

t = np.linspace(0, 8*np.pi,1000)

x = np.cos(k*t) * np.cos(t)
y = np.cos(k*t) * np.sin(t)

plt.plot(x,y, 'm', linewidth=3)
plt.axis('square')
plt.axis('off')
plt.show()
```

```python
fig, ax = plt.subplots(3,3, figsize=(10,6))
row, col = np.indices((3,3))
                      
                      
                      
                      
for i in range(0,9):  ## 
    
    k =i/2 ##i have specified k to be a looping index divided by 2
    
    x = np.cos(k*t) * np.cos(t)
    y = np.cos(k*t) * np.sin(t)
    
    r = row.ravel()[i]
    c = col.ravel()[i]

    ax[r,c].plot(x,y, 'k')
    ax[r,c].axis('square')
    ax[r,c].axis('off')
    ax[r,c].set_title('k = %s' %k)  ## %s means substitution
    
    
plt.show()
    
    
  
```
